# 1. Olvass be öt darab tetszőleges, 32x32 pixeles, színes képet, jelenítsd meg és írd ki az RGB csatornákat tartalmazó tömbök tartalmát!

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage as ndimage
%matplotlib inline
```

```python
#kepek beolvasasa egy tombbe
imgs=np.empty([5,32,32,3]) #5 kep, 32x32 pixel, 3 csatorna
for i,fname in enumerate(["img1.bmp","img2.bmp","img3.bmp","img4.bmp","img5.bmp"]):
    imgs[i]=ndimage.imread(fname)
```

```python
#kiiras
#np.set_printoptions(threshold=np.inf) #igy az osszes elemet kiirja omlesztve
for img in range(0,5):
    for rgb in range(0,3):
        print(str(img+1)+". kep "+str(rgb+1)+". csatorna:\n"+str(imgs[img,:,:,rgb])+"\n")
```

```python
#kirajzolas
for img in range(0,5):
    plt.figure()
    for rgb in range(0,3):
        plt.subplot(1,3,rgb+1)
        plt.imshow(imgs[img,:,:,rgb],cmap=plt.cm.gray)
        plt.axis("off")
        plt.title(str(img+1)+". kep "+str(rgb+1)+". csat.")
```

# 2. Csatornánként számítsd ki az egyes pixelek átlagát és szórását az öt képre, majd alakítsd át ezeket 0 várható értékű, 1 szórasú adathalmazzá!

```python
avgs=np.empty([5,3]) #atlagok
stdevs=np.empty([5,3]) #szorasok
for img in range(0,5):
    for rgb in range(0,3):
        avgs[img,rgb]=imgs[img,:,:,rgb].mean()
        stdevs[img,rgb]=imgs[img,:,:,rgb].std()
```

```python
#kiiras
for img in range(0,5):
    for rgb in range(0,3):
        print("%d. kep %d. csatorna atlaga: %.2f, szorasa: %.2f" %
              (img+1, rgb+1, avgs[img,rgb], stdevs[img,rgb]))
```

```python
#standardizalas
for img in range(0,5):
    for rgb in range(0,3):
        imgs[img,:,:,rgb]-=avgs[img,rgb]
        imgs[img,:,:,rgb]/=stdevs[img,rgb]
```

```python
#ellenorzes
for img in range(0,5):
    for rgb in range(0,3):
        print("%d. kep %d. csatorna atlaga: %.2f, szorasa: %.2f" %
              (img+1, rgb+1, imgs[img,:,:,rgb].mean(), imgs[img,:,:,rgb].std()))
```

# 3. Olvass be egy tetszőleges hangfájlt, és jelenítsd meg a spektrogramját!

```python
import scipy.io.wavfile as wav
```

```python
fs,x=wav.read("sine_fm.wav") #frekvenciamodulalt szinusz (880Hz vivo, 200Hz loket, 1Hz modulacio)
fs #mintaveteli frekvencia
```

```python
#a jel egy darabjanak kirajzolasa
t=np.linspace(0,np.floor(x.size/fs),x.size)
plt.plot(t,x)
plt.xlim([0,0.02]);
```

```python
#spektrogram 1024 pontos FFT-vel, Hanning-ablakkal
plt.specgram(x,1024,fs);
```

# 4. Olvass be egy tetszőleges angol nyelvű szövegfájlt, és hisztogramon jelenítsd meg a betűk előfordulásának gyakoriságát!

```python
file=open("shakespeare.txt","r") #William Shakespeare: A Lover's Complaint
txt=file.read()
```

```python
d={} #ures szotar
for c in txt:
    if (ord(c) in range(ord('A'),ord('Z')+1)) or (ord(c) in range(ord('a'),ord('z')+1)): #csak betuk
        if c in d:
            d[c]+=1
        else:
            d[c]=1
d
```

```python
#rendezes
x=list(sorted(d.keys()))
y=np.empty(len(d),dtype=int)
for i,key in enumerate(x):
    y[i]=d[key]
```

```python
#abrazolas
plt.xticks(range(len(x)),x);
plt.bar(range(len(x)),y);
```
