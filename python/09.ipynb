```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
import sklearn.metrics as metrics
```

```python
def f(x): return np.sqrt(np.log(x))+0.2*np.sin(x); # a megtanulando fuggveny
x_train=np.random.uniform(1,10,100); # tanitopontok
x_pred=np.arange(1,10,0.01); # ezekben a pontokban szeretnenk kozeliteni a fuggvenyt
```

```python
# tanitopontok
y_train=f(x_train);
plt.figure(1);
plt.plot(x_train,y_train,'.');
plt.title('Tanitopontok');
plt.xlabel('x');
plt.ylabel('y');
# random forest
rfr=RandomForestRegressor(n_estimators=30,n_jobs=-1);
rfr.fit(x_train.reshape(-1,1),y_train);
y_pred=rfr.predict(x_pred.reshape(-1,1));
plt.figure(2,figsize=[12,5]);
plt.subplot(1,2,1);
plt.plot(x_pred,y_pred,'b.',x_pred,f(x_pred),'r');
plt.title('Regresszio (random forest)');
plt.xlabel('x');
plt.ylabel('y');
plt.subplot(1,2,2);
plt.plot(x_pred,y_pred-f(x_pred),'r');
plt.title('Hiba (random forest)');
plt.xlabel('x');
plt.ylabel('e');
print('Random forest R^2='+str(metrics.r2_score(y_pred,f(x_pred))))
# gradient boost
gbr=GradientBoostingRegressor();
gbr.fit(x_train.reshape(-1,1),y_train);
y_pred=gbr.predict(x_pred.reshape(-1,1));
plt.figure(3,figsize=[12,5]);
plt.subplot(1,2,1);
plt.plot(x_pred,y_pred,'b.',x_pred,f(x_pred),'r');
plt.title('Regresszio (gradient boost)');
plt.xlabel('x');
plt.ylabel('y');
plt.subplot(1,2,2);
plt.plot(x_pred,y_pred-f(x_pred),'r');
plt.title('Hiba (gradient boost)');
plt.xlabel('x');
plt.ylabel('e');
print('Gradient boost R^2='+str(metrics.r2_score(y_pred,f(x_pred))))
```
